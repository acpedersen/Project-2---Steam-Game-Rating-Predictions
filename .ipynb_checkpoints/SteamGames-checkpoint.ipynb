```python
# import libraries
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, precision_score, f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.model_selection import train_test_split
```

```python
initialDf = pd.read_csv("steam.csv")
initialDf.head()
```

```python
for var in initialDf.columns:
    if len(initialDf[var]) != len(initialDf[var].dropna()):
        print(var)
print('Done')
#Check to see if there are any na's
```

```python
df = pd.DataFrame()
df["Name"] = initialDf["name"].apply(lambda x: str(x))
df["ReleaseYear"] = initialDf["release_date"].apply(lambda x: int(x[0:4]))
df["ReleaseMonth"] = initialDf["release_date"].apply(lambda x: int(x[5:7]))
df["English"] = initialDf["english"].apply(lambda x: int(x))
df["Developer"] = initialDf["developer"].apply(lambda x: str(x))
df["Publisher"] = initialDf["publisher"].apply(lambda x: str(x))
df["Windows"] = initialDf["platforms"].apply(lambda x: bool("windows" in x))
df["Mac"] = initialDf["platforms"].apply(lambda x: bool("mac" in x))
df["Linux"] = initialDf["platforms"].apply(lambda x: bool("linux" in x))
df["RequiredAge"] = initialDf["required_age"].apply(lambda x: int(x))
df["Categories"] = initialDf["categories"].apply(lambda x: str(x))
df["Genre"] = initialDf["genres"].apply(lambda x: str(x))
df["Price"] = initialDf["price"].apply(lambda x: float(x))


#df["PositiveRatings"] = initialDf["positive_ratings"].apply(lambda x: int(x))
#df["NegativeRatings"] = initialDf["negative_ratings"].apply(lambda x: int(x))
df["MedianPlaytime"] = initialDf["median_playtime"].apply(lambda x: int(x))
df["AveragePlaytime"] = initialDf["average_playtime"].apply(lambda x: int(x))
df["LowerOwners"] = initialDf["owners"].apply(lambda x: int(x[:x.find("-")]))
df["UpperOwners"] = initialDf["owners"].apply(lambda x: int(x[x.find("-")+1:]))
```

```python
#Normalize the data to make it easier to change
initialNormalizeFeatures = ['positive_ratings', 'negative_ratings']
NormalizeFeatures = ['PositiveRatings', 'NegativeRatings']
dfNormalize = pd.DataFrame(data=MinMaxScaler().fit_transform(initialDf[initialNormalizeFeatures]), columns=NormalizeFeatures)

df = pd.concat([df, dfNormalize], axis=1)
df.head()
```

```python
#Cant have any 0's if we are going to divide. Shouldn't skew the predictability of the algorithm
df["NegativeRatings"].replace({0: 1}, inplace=True)
```

```python
#Created two new variables to predict as the positive and negative ratings are a rising number and not easily predicted
#So instead we created a ratio that is comparable
df['Pos_over_Neg'] =  (df['PositiveRatings'] / df['NegativeRatings']).apply(lambda x: float(x))

#As well as a boolean value so we can use logistic regression as the project specifies
df['Ratio_over_std'] = np.where(df['Pos_over_Neg']>=df['Pos_over_Neg'].std(), 1, 0)
```

```python
#set the data for the specific types of data to 1 if true or zero if false
#helps with using the data for further use
for col in df.select_dtypes(include=['bool']).columns:
    df[col] = df[col].apply(lambda x: int(1 if x else 0))
#for col in df.select_dtypes(include=['int']).columns:
#    df[col] = df[col].apply(lambda x: float(x))
df.head()
```

```python
#Creating a list of all features to use in prediction
features = ['English', 'Windows', 'Mac', 'Linux', 'RequiredAge', 'Price', 'MedianPlaytime', 'AveragePlaytime']


#Our first target, (Ratio of Positive Ratings/ Negative ratings) over the std or not
target = 'Ratio_over_std'
#Our second target, (Ratio of Positive Ratings/ Negative ratings)
secTarget = 'Pos_over_Neg'
```

```python
#Normalize the two float sets and convert them to tensors
tfFeatures = tf.keras.utils.normalize(df[features].values)
tfTarget = tf.constant(df[target].values, tf.bool)
tfSecTarget = tf.keras.utils.normalize(df[secTarget].values)
```

```python
#Shuffle them for a random selection of the train, val, and test
#Also reshape the targets so they can be properly input into the model
tf.random.set_random_seed(42)
tfFeatures = tf.random.shuffle(tfFeatures)
tfTarget = tf.reshape(tf.random.shuffle(tfTarget), [tf.size(tfTarget), 1])
tfSecTarget = tf.reshape(tf.random.shuffle(tfSecTarget), [tf.size(tfSecTarget), 1])
```

```python
print(tfFeatures)
print(tfTarget)
print(tfSecTarget)
```

```python
#Split the data
lenTrain, lenVal, lenTest = 25000, 1000, 1075
X_train, X_val, X_test = tf.split(tfFeatures, [lenTrain, lenVal, lenTest], axis=0)
y_train, y_val, y_test = tf.split(tfTarget, [lenTrain, lenVal, lenTest], axis=0)
y2_train, y2_val, y2_test = tf.split(tfSecTarget, [lenTrain, lenVal, lenTest], axis=0)
```

```python
#Create a model with 4 layers
afunc = tf.nn.relu

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(8, activation=afunc))
model.add(tf.keras.layers.Dense(10, activation=afunc))
model.add(tf.keras.layers.Dense(3, activation=afunc))
model.add(tf.keras.layers.Dense(7, activation=afunc))
#Last layer is the prediction layer
model.add(tf.keras.layers.Dense(1, activation=tf.nn.sigmoid))
```

```python
#Using BinaryCrossentropy and BinaryAccuracy for this model
optim = 'adam'
model.compile(optimizer=optim,
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=[tf.keras.metrics.BinaryAccuracy()])
```

```python
#Input the training data and the validation data
#Also save the history for looking at the model over epochs
hist = model.fit(X_train, y_train, steps_per_epoch=5, epochs=15, batch_size=5000, validation_data = (X_val, y_val), validation_steps=5)
```

```python
#One thing we don't understand is why the model accuracy plateaus at 85%
#While the loss continues down at a steady rate
plt.plot(hist.history['binary_accuracy'])
plt.plot(hist.history['val_binary_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
#Checking against the test set to ensure there is no underfitting or overfitting
val_loss, val_acc = model.evaluate(X_test, y_test, steps=1)
print(val_loss)
print(val_acc)
```

```python
#Second model for predicting the value of the pos_over_neg
afunc = tf.nn.relu

model2 = tf.keras.models.Sequential()
model2.add(tf.keras.layers.Flatten())

model2.add(tf.keras.layers.Dense(8, activation=afunc))
model2.add(tf.keras.layers.Dense(10, activation=afunc))
model2.add(tf.keras.layers.Dense(3, activation=afunc))
model2.add(tf.keras.layers.Dense(7, activation=afunc))

model2.add(tf.keras.layers.Dense(1))
```

```python
model2.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.MeanSquaredError(),
              metrics=[tf.keras.metrics.MeanAbsoluteError()])
```

```python
#Same model fitting as the first one
hist2 = model2.fit(X_train, y2_train, steps_per_epoch=5, epochs=100, batch_size=5000, validation_data = (X_val, y2_val), validation_steps=5)
```
